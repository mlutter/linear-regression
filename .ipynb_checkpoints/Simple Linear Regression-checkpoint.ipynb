# Simple Linear Regression

This is a linear regression being run on a single feature of the default Boston housing dataset. This uses a train/test split and charts the results. 

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np

boston = datasets.load_boston()

X = boston.data[:, np.newaxis, 5] #this takes just the RM feature. 
y = boston.target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

reg_all = linear_model.LinearRegression()
reg_all.fit(X_train, y_train)

y_pred = reg_all.predict(X_test)

print("Regression score: {}".format(reg_all.score(X_test, y_test)))

# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, reg_all.predict(X_train), color = 'blue')
plt.title('(Train) Average number of rooms per dwelling vs Median Value')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()

# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, reg_all.predict(X_train), color = 'blue')
plt.title('(Test) Average number of rooms per dwelling vs Median Value')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```
